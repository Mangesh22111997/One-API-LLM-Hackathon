```python

from intel_extension_for_transformers.neural_chat import PipelineConfig
from intel_extension_for_transformers.neural_chat import plugins
plugins.retrieval.enable=True
plugins.retrieval.args["input_path"]="./Indian Constitution.pdf"
config = PipelineConfig(plugins=plugins)
from intel_extension_for_transformers.neural_chat import build_chatbot
chatbot = build_chatbot(config)
```

```python
def chatbot_intel(question):
  response = chatbot.predict(question)
  return response
```

```python
from ipywidgets import widgets
from IPython.display import display, HTML, clear_output

# Function to get response based on user input
def get_response(user_input):
    # Add your custom logic here to generate a response
    return f"Response: {user_input}"

# Create interactive widgets
input_text = widgets.Text(placeholder="Type here...", layout=widgets.Layout(width='50%'))
button = widgets.Button(description="Get Response", button_style='success')
output_text = widgets.Output()

# Function to handle button click
def on_button_click(b):
    user_input = input_text.value
    response = chatbot_intel(user_input)

    with output_text:
        clear_output(wait=True)
        print(response)

# Set button click event
button.on_click(on_button_click)

# Display the interactive widgets
display(HTML("<h1 style='color: #FF6600;'>Bharat LawLex </h1><h3 style='color: #3498db;'>by InfyTechies </h3>"))
display(input_text, button, output_text)
```
